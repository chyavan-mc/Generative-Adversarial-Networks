```python
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd
```

```python
df = pd.read_csv("gifs_to_form.csv")
df
```

```python
import os

for j in range(len(df)):
    s = pd.DataFrame(eval(df.iloc[j]["drawing"]))
    x_min, x_max = s.loc[:,0].map(min).min(), s.loc[:,0].map(max).max()
    y_min, y_max = s.loc[:,1].map(min).min(), s.loc[:,1].map(max).max()

    for row_idx in range(len(s)):
        rows = s.iloc[:row_idx+1]
        fig = plt.figure(figsize=((x_max - x_min)/100, (y_max - y_min)/100))
        for row_iter in range(len(rows)):
            row = rows.iloc[row_iter]
            for i in range(1, len(row[0])):
                for k in range(1,i):
                    plt.plot([(row[0][k-1]-x_min)/100, (row[0][k]-x_min)/100], [(row[1][k-1]-y_min)/100, (row[1][k]-y_min)/100], linestyle='-', color="black")
                plt.axis('off')
                plt.xlim(0, (x_max-x_min)/100)
                plt.ylim(0, (y_max-y_min)/100)
                if not os.path.exists(f"./bat_images/bat_{j}"):
                    os.makedirs(f"./bat_images/bat_{j}")
                plt.savefig(f"./bat_images/bat_{j}/frame_{row_iter}_{i}.jpg")
```

```python
import glob
import imageio
```

```python
for i in range(6):
    anim_file = f'./bat_{i}.gif'

    with imageio.get_writer(anim_file, mode='I') as writer:
      filenames = glob.glob(f'./bat_images/bat_{i}/frame*.jpg')
      filenames = sorted(filenames)
      for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
      image = imageio.imread(filename)
      writer.append_data(image)
```
